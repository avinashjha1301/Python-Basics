**Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples**

Ansswer:- Anomaly Detection is a technique in machine learning and data analysis used to identify unusual patterns, behaviors, or observations that differ significantly from normal data. These unusual observations are called anomalies or outliers and may indicate errors, fraud, rare events, or system failures.

There are three main types of anomalies:

Point Anomalies
A point anomaly occurs when a single data point is abnormal compared to the rest of the data.
Example: In a bank transaction dataset, a sudden transfer of ‚Çπ10,00,000 from an account that usually transfers ‚Çπ5,000 is a point anomaly.

Contextual Anomalies
A contextual anomaly is abnormal only in a specific context, such as time, location, or condition.
Example: A temperature of 30¬∞C may be normal in summer but anomalous in winter. Similarly, high website traffic at midnight might be considered abnormal.

Collective Anomalies
A collective anomaly occurs when a group of related data points is unusual, even if individual points seem normal.
Example: In network monitoring, a sequence of small data packets could collectively indicate a cyberattack, even though each packet appears normal.

Anomaly detection is widely used in fraud detection, healthcare monitoring, cybersecurity, and predictive maintenance.

**Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases**

Answer:- Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are popular anomaly detection methods, but they differ in approach and use cases.

Isolation Forest is a tree-based algorithm that detects anomalies by randomly partitioning data. Since anomalies are rare and different, they get isolated in fewer splits, resulting in shorter path lengths. It works well for high-dimensional datasets and large-scale problems such as fraud detection and intrusion detection. It is efficient and does not require distance calculations.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies anomalies as points that do not belong to any dense cluster. It groups data based on density and labels sparse points as outliers. DBSCAN is effective for datasets with clusters of arbitrary shapes and is widely used in spatial data analysis, geolocation data, and image processing. However, it struggles with varying density clusters.

Local Outlier Factor (LOF) is a density-based method that compares the local density of a point with its neighbors. Points with significantly lower density are considered anomalies. LOF is best for detecting local outliers in datasets where anomalies occur in specific regions, such as network traffic analysis or medical diagnosis.

**Question 3: What are the key components of a Time Series? Explain each with one example.**

Answer:- A time series is a sequence of data points collected or recorded at regular time intervals. Time series analysis helps identify patterns and make future predictions. The key components of a time series are:

Trend
The trend represents the long-term movement or direction in the data over time.
Example: The steady increase in global smartphone sales over the past 10 years shows an upward trend.

Seasonality
Seasonality refers to repeating patterns or fluctuations that occur at fixed intervals, such as daily, monthly, or yearly cycles.
Example: Ice cream sales increase every summer and decrease in winter, showing seasonal behavior.

Cyclical Component
Cyclical patterns are long-term fluctuations that occur over extended periods but do not follow a fixed schedule like seasonality.
Example: Economic growth and recession cycles affect stock market performance over several years.

Irregular or Random Component (Noise)
This component consists of unpredictable variations caused by sudden or unexpected events.
Example: A sudden drop in airline bookings due to a natural disaster is an irregular pattern.

Level
The level is the average value around which the time series fluctuates.
Example: The consistent average daily electricity usage in a city represents its level.

**Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**

Answer:- A stationary time series is one whose statistical properties such as mean, variance, and autocorrelation remain constant over time. In other words, the data does not show trends or changing patterns, making it easier to model and forecast.

A series is non-stationary if it has trends, seasonality, or changing variance. Most real-world data like stock prices or sales are non-stationary.

Testing Stationarity

Common methods to test stationarity include:

Augmented Dickey-Fuller (ADF) Test
It checks whether a unit root exists. If the p-value is less than 0.05, the series is likely stationary.

KPSS Test
It tests the opposite hypothesis. A high p-value indicates stationarity.

Visual Inspection
Plotting the series helps detect trends and seasonality.

Transforming Non-Stationary to Stationary

To make a series stationary, common transformations include:

Differencing
Subtract the previous value from the current value to remove trend.

**Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Simple sample taxi fare dataset
X = np.array([
    [10, 2], [12, 3], [15, 2.5], [14, 3.2], [11, 2.8],   # normal trips
    [80, 15], [90, 18]                                   # anomalies
])

# Isolation Forest model
model = IsolationForest(contamination=0.2, random_state=0)
y_pred = model.fit_predict(X)

# Separate normal and anomalies
normal = X[y_pred == 1]
anomaly = X[y_pred == -1]

# Scatter plot
plt.scatter(normal[:,0], normal[:,1], label="Normal")
plt.scatter(anomaly[:,0], anomaly[:,1], label="Anomaly")

plt.xlabel("Fare Amount")
plt.ylabel("Trip Distance")
plt.title("Isolation Forest Anomaly Detection")
plt.legend()
plt.show()
```

**Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.**

**Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# -----------------------------
# Sample Numerical Dataset
# -----------------------------
np.random.seed(42)

# Normal data points
X_normal = np.random.normal(loc=10, scale=2, size=(200, 2))

# Add anomalies (outliers)
X_outliers = np.random.normal(loc=25, scale=1, size=(10, 2))

# Combine dataset
X = np.vstack((X_normal, X_outliers))

# -----------------------------
# Apply Local Outlier Factor
# -----------------------------
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)

# Separate normal vs anomalies
normal = X[y_pred == 1]
anomaly = X[y_pred == -1]

# -----------------------------
# Visualization
# -----------------------------
plt.figure(figsize=(8,6))
plt.scatter(normal[:,0], normal[:,1], label="Normal Points")
plt.scatter(anomaly[:,0], anomaly[:,1], label="Anomalies (LOF)")

plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

**Question 10: You are working as a data scientist for a power grid monitoring company.**
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:
‚óè How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?
‚óè Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?
‚óè How would you validate and monitor the performance over time?
‚óè How would this solution help business decisions or operations?


Answer:- Real-Time Data Science Workflow (Power Grid Monitoring)
1. Anomaly Detection in Streaming Data

To detect abnormal spikes or drops every 15 minutes:

Isolation Forest: Best for real-time because it is fast and works well on high-dimensional features.

LOF (Local Outlier Factor): Detects anomalies based on local density, useful when unusual behavior happens only in specific regions.

DBSCAN: Detects anomalies as noise points, but not ideal for streaming because it is slower.

Recommended: Isolation Forest for real-time anomaly alerts

2. Short-Term Forecasting Model

Energy usage depends on:

time of day

seasonality

weather conditions

So the best model is:

SARIMAX (Seasonal ARIMA + external variables like temperature, humidity)

3. Validation & Monitoring

Use rolling window validation

Track forecast errors: MAE, RMSE

Monitor anomaly counts over time

Retrain models weekly/monthly

4. Business Impact

This solution helps in:

preventing blackouts

detecting equipment failure early

load balancing and operational planning

improving customer reliability

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Sample 15-min energy data
t = pd.date_range("2025-01-01", periods=100, freq="15min")
y = np.random.normal(100, 10, 100)
y[30], y[70] = 180, 40   # anomalies

df = pd.DataFrame({"energy": y}, index=t)

# Anomaly Detection (Isolation Forest)
iso = IsolationForest(contamination=0.05)
df["anomaly"] = iso.fit_predict(df[["energy"]])

# Forecasting (SARIMA/SARIMAX)
model = SARIMAX(df["energy"], order=(1,1,1), seasonal_order=(1,1,1,4))
fit = model.fit(disp=False)
forecast = fit.forecast(steps=10)

# Plot
plt.plot(df["energy"], label="Energy")
plt.scatter(df[df["anomaly"]==-1].index,
            df[df["anomaly"]==-1]["energy"], label="Anomaly")

plt.plot(forecast.index, forecast, label="Forecast")
plt.legend()
plt.show()
```
