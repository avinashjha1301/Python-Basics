**Question 1 : Explain the difference between Absolute, Relative, and Mixed Cell Referencing in Excel with examples.**

Answer:- In Microsoft Excel, cell referencing determines how formulas behave when copied to other cells. There are three main types: relative, absolute, and mixed referencing.

Relative referencing is the default type in Excel. In this, cell references change automatically when a formula is copied to another cell. For example, if a formula =A1+B1 is copied from row 1 to row 2, it becomes =A2+B2. This is useful when performing the same calculation across multiple rows or columns.

Absolute referencing, on the other hand, keeps the cell reference fixed regardless of where the formula is copied. It uses the dollar sign ($) before the column and row, such as =$A$1. For example, in the formula =A1*$B$1, the value in cell B1 remains constant. This is commonly used when applying a fixed value like a tax rate or constant multiplier.

Mixed referencing is a combination of both relative and absolute referencing. In this, either the row or the column is fixed, but not both. For example, $A1 keeps the column constant while allowing the row to change, and A$1 keeps the row constant while allowing the column to change. Mixed referencing is especially useful in creating tables like multiplication tables where one dimension needs to remain fixed.

**Question 3 : What are Text Functions in Excel? Mention any five with examples.**

Answer :- In Microsoft Excel, **Text Functions** are used to manipulate and manage text data such as names, words, or sentences. These functions help in combining, extracting, formatting, or changing text easily.

Five commonly used text functions are:

1. **LEFT()** – Extracts characters from the beginning of a text.
   Example: `=LEFT("Excel",2)` returns **"Ex"**.

2. **RIGHT()** – Extracts characters from the end of a text.
   Example: `=RIGHT("Excel",2)` returns **"el"**.

3. **MID()** – Extracts characters from the middle of a text.
   Example: `=MID("Excel",2,3)` returns **"xce"**.

4. **UPPER()** – Converts text into uppercase.
   Example: `=UPPER("excel")` returns **"EXCEL"**.

5. **LOWER()** – Converts text into lowercase.
   Example: `=LOWER("EXCEL")` returns **"excel"**.

These functions are very helpful in cleaning and organizing text data efficiently.


**Question 4 : What is the use of Scenario Manager in decision making?**

Answer :- In Microsoft Excel, the **Scenario Manager** is a tool used for *what-if analysis* that helps in decision making by comparing different sets of input values and their outcomes. It allows users to create and save multiple scenarios, such as best case, worst case, and most likely case, within the same worksheet.

The main use of Scenario Manager is to analyze how changes in key variables affect the final result. **Question 5 : Define the purpose of VLOOKUP and HLOOKUP. How are they different from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?**

Answer:- In Microsoft Excel, lookup functions are used to search for data in a table and return corresponding values.

**VLOOKUP (Vertical Lookup)** is used to search for a value in the first column of a table and return a corresponding value from another column in the same row. For example, it can find a student’s marks based on their roll number.

**HLOOKUP (Horizontal Lookup)** works similarly but searches for a value in the first row of a table and returns a value from a specified row below. It is useful when data is arranged horizontally.

**XLOOKUP** is a more advanced and flexible function that overcomes the limitations of VLOOKUP and HLOOKUP. It can search both vertically and horizontally, does not require a fixed column or row index, and can return values from any direction (left, right, up, or down). It also handles errors more effectively.

The main difference is that VLOOKUP and HLOOKUP are limited in flexibility and require specific data structures, while XLOOKUP is more powerful and easier to use.

Between **XLOOKUP and INDEX-MATCH**, XLOOKUP is generally considered better for most users because it is simpler, more readable, and combines multiple features in one function. However, INDEX-MATCH is still useful in older Excel versions where XLOOKUP is not available and is known for its flexibility and efficiency in complex datasets.
