**Question 2:** Explain overfitting and underfitting in ML. How can you detect and prevent
them?

Overfitting and Underfitting in Machine Learning
1. Overfitting

Definition:
Overfitting occurs when a model learns the training data too well, including noise and unnecessary patterns.
As a result, it performs well on training data but poorly on new, unseen data.

Signs of overfitting:

Very low training error

High validation/test error

Cause:
The model is too complex (high variance).

2. Underfitting

Definition:
Underfitting occurs when the model is too simple to learn important patterns in the data.
It performs poorly on both training and test data.

Signs of underfitting:

High training error

High validation/test error

Cause:
The model has too little complexity (high bias).

Bias‚ÄìVariance Tradeoff

Underfitting = High bias, low variance

Overfitting = Low bias, high variance

Good ML models find a balance between the two, minimizing total error.

How to Detect Overfitting and Underfitting
Detect Overfitting

Training accuracy very high, validation much lower

Validation loss decreases initially then increases (after some epochs in neural networks)

Large gap between train and test performance

Detect Underfitting

Training accuracy is low

Model fails to improve with more training

Training and validation errors both high

How to Prevent Overfitting
1. Cross-Validation

Use k-fold cross-validation to ensure model performs consistently on different data slices.

Helps detect overfitting early and choose the best model/hyperparameters.

2. Regularization

Apply penalties to limit model complexity.

L1 (Lasso) ‚Üí reduces feature count

L2 (Ridge) ‚Üí shrinks coefficients

Elastic Net ‚Üí combines both

Prevents the model from relying on noisy features.

3. Reduce Model Complexity

Use simpler models (e.g., reduce tree depth, fewer layers, fewer parameters).

4. Early Stopping (for neural networks)

Stop training when validation error starts increasing.

5. Data Augmentation / Collect More Data

Helps the model generalize better.

How to Prevent Underfitting

Increase model complexity (more layers, deeper trees, more features).

Reduce regularization strength.

Train for more epochs or tune learning rate.

Add more relevant features.

Handling Missing Values in a Dataset

Missing values are a common issue in real-world datasets and must be handled properly to avoid biased or incorrect model results. There are several methods depending on the type of data and the extent of missingness.

1. Deletion Methods
a) Listwise Deletion (Remove entire row)

Remove rows that contain any missing values.

Useful when missing data is very small (<5%) and appears randomly.

Example:
Dataset:

Age	Salary
25	40,000
‚Äî	50,000
30	45,000

Row 2 is removed because "Age" is missing.

Pros: Simple, fast.
Cons: Loss of data, not good if missing values are many.

b) Column Deletion

Remove entire columns if the feature has too many missing values (>60‚Äì70%).

Example:
A column like ‚ÄúAddress Line 2‚Äù with 80% missing values can be removed.

2. Imputation Using Statistical Methods
a) Mean/Median/Mode Imputation

Replace missing values with statistical summaries.

Mean ‚Üí for continuous & normal data
Median ‚Üí for skewed numerical data
Mode ‚Üí for categorical data

Example:
Age column: [25, 30, ‚Äî, 35]
Mean age = (25 + 30 + 35) / 3 = 30
Missing value is replaced with 30.

Pros: Easy, maintains dataset size.
Cons: Reduces variance & may introduce bias.

3. Predictive Modeling (Advanced Imputation)

Use machine learning models to predict missing values.

a) Regression Imputation

Build a regression model using complete data and predict missing values.

Example:
Predict missing "Salary" using Age, Experience, Education.

b) KNN Imputation

Replace missing values with the average value from k nearest neighbors.

Example:
If Age is missing, look at 5 nearest neighbors and take their mean age.

c) MICE (Multiple Imputation by Chained Equations)

Iteratively predicts missing values using multiple models.

Pros: More accurate, preserves relationships in data.
Cons: More computationally expensive.

**Question 4:**What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).


An imbalanced dataset is a dataset where the number of samples in one class is much higher than in the other classes.
This causes machine learning models to become biased toward the majority class and perform poorly on minority classes.

Example:
Fraud detection dataset:

99% transactions = Not Fraud

1% transactions = Fraud

A model may predict ‚ÄúNot Fraud‚Äù every time and still get 99% accuracy‚Äîbut it is useless.

Techniques to Handle Imbalanced Data
1. Oversampling the Minority Class

Oversampling increases the number of minority samples so the dataset becomes more balanced.

A) Random Oversampling (Simple Method)

Theory:
Randomly duplicate existing minority class samples until the classes are balanced.

Practical Example:
If class distribution is:

Class 0: 900 samples

Class 1: 100 samples

Random oversampling creates additional copies of Class 1 until both reach 900.

Pros:
Simple and effective for small datasets.

Cons:
Overfitting risk, because duplicated data may repeat the same patterns.

B) SMOTE (Synthetic Minority Oversampling Technique)

Theory:
Instead of duplicating data, SMOTE creates synthetic (new) minority-class examples by interpolating between nearest neighbors.

How it works:
For each minority example:

Choose k nearest neighbors

Generate new samples between them using linear interpolation

Practical Example:
If "Fraud" class has only 100 samples, SMOTE creates additional synthetic Fraud samples such as:
new_sample = sample + random(0,1) * (neighbor_sample - sample)

Pros:

Reduces overfitting

Produces more realistic samples

Cons:

Can create noisy synthetic points

Should be used only on numerical data

2. Undersampling the Majority Class

Reduces the number of majority samples to match the minority.

A) Random Undersampling

Theory:
Randomly remove samples from the majority class to balance the dataset.

Practical Example:
Class 0 = 900
Class 1 = 100
Remove 800 samples from Class 0 ‚Üí Both become 100.

Pros:

Fast and simple

Reduces training time

Cons:

Risk of losing important information

3. Using Class Weights in Models

Theory:
Assign higher penalty/weight to the minority class so the model focuses more on it.

Practical Example:
In models like Logistic Regression, SVM, Random Forest:
class_weight = {0:1, 1:10}
‚Üí Misclassifying class 1 (minority) becomes 10x more costly.

Pros:

No data removal or duplication

Commonly used in real projects

Cons:

Requires tuning

Might cause instability in small datasets

Why Feature Scaling is Important in ML

Feature scaling is the process of transforming numerical features so that they are on a similar scale.
It is essential because many machine learning algorithms are sensitive to the magnitude of features.

Why scaling matters:
1. Distance-based algorithms

Algorithms like:

KNN

K-Means

SVM with RBF kernel

depend on distance calculations (Euclidean distance).
If one feature has a large range (e.g., income = 0‚Äì10,00,000) and another has a small range (age = 0‚Äì100), the large-range feature will dominate the distance metric.

Without scaling:
Model gives unfair weight to large-value features.

2. Gradient Descent Optimization

Models like:

Linear Regression

Logistic Regression

Neural Networks

use gradient descent to minimize loss.

If features vary greatly in scale, gradients can become unstable ‚Üí

Slow convergence

Zig-zagging during training

Risk of getting stuck in local minima

Scaling improves training speed and stability.

**Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?**  


Label Encoding vs One-Hot Encoding
1. Label Encoding

Definition:
Converts each category into a numeric label (0, 1, 2, 3...).

Example:
Color = {Red, Blue, Green}
‚Üí Red = 0, Blue = 1, Green = 2

When to Use:

When the categorical data has a natural order (ordinal)
Examples:

Size: Small < Medium < Large

Education: High School < Bachelor‚Äôs < Master‚Äôs < PhD

Rating: Poor < Average < Good < Excellent

Advantages:

Simple and compact

Works well with tree-based models (Decision Tree, Random Forest, XGBoost) because they split, not compute distance

Disadvantages:

Implies ordinal relationship even when categories are nominal

Can mislead models like KNN, Linear Regression, SVM, where numeric values create false importance

2. One-Hot Encoding

Definition:
Creates binary columns (0/1) for each category.

Example:
Color = {Red, Blue, Green}

Red	Blue	Green
1	0	0
When to Use:

When categories are nominal (no natural order)
Examples:

Gender: Male, Female

Cities: Delhi, Mumbai, Chennai

Product Type: Electronics, Clothing, Furniture

Advantages:

No false ordinal relationship

Works well with distance-based and linear models
(KNN, Logistic Regression, SVM, Neural Networks)

Disadvantages:

Increases dimensionality ‚Üí "curse of dimensionality"

Not suitable when categories are very large (e.g., zip codes)


```python
'''Question 7: Google Play Store Dataset a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?'''

import pandas as pd

df = pd.read_csv("googleplaystore.csv")

print(df[['Category', 'Rating']].info())
print(df[['Category', 'Rating']].head())

df_clean = df.dropna(subset=['Rating'])

avg_rating = df_clean.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print("\nAverage rating by category (top 20):\n", avg_rating.head(20))
print("\nLowest rating categories (bottom 20):\n", avg_rating.tail(20))

count_by_cat = df_clean['Category'].value_counts().sort_values(ascending=False)
print("\nNumber of apps per category (top 10):\n", count_by_cat.head(10))

summary = pd.DataFrame({
    'num_apps': count_by_cat,
    'avg_rating': avg_rating
}).dropna().sort_values(by='avg_rating', ascending=False)

print("\nSummary (category, # apps, avg rating):\n", summary.head(15))
```
