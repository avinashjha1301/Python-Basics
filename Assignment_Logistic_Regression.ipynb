**Question 2:** **What are the key assumptions of Simple Linear Regression?**

**Answer:**  --> The key assumptions of Simple Linear Regression (SLR) are conditions that must hold true for the model to produce valid, unbiased, and reliable results. These assumptions are:

Linearity
There is a linear relationship between the independent variable (X) and the dependent variable (Y). The change in Y is proportional to the change in X.

Independence of Errors
The residuals (errors) are independent of each other. This means the error for one observation does not influence another, which is especially important in time-series data.

Homoscedasticity
The variance of residuals is constant across all values of X. If the spread of errors increases or decreases with X, heteroscedasticity occurs.

Normality of Errors
The residuals are normally distributed. This assumption is mainly required for valid hypothesis testing and confidence interval estimation.

No Significant Outliers
Extreme values should not disproportionately influence the regression line, as outliers can distort the slope and intercept.

No Measurement Error in X
The independent variable is assumed to be measured accurately without error.

**Question 5: What is the method of least squares in linear regression?**

**Answer:** --> The method of least squares in linear regression is a technique used to estimate the best-fitting regression line for a given dataset. It works by minimizing the sum of the squares of the residuals, where residuals are the differences between the observed values and the predicted values. Squaring the residuals ensures that positive and negative errors do not cancel each other out and gives more weight to larger errors. By finding the values of the intercept and slope that minimize this total squared error, the method produces the most accurate linear relationship between the independent and dependent variables.


**Question 6: What is Logistic Regression? How does it differ from Linear Regression?**


**Answer:** --> Logistic Regression is a statistical and machine learning technique used to predict binary outcomes (such as yes/no, 0/1, pass/fail) by modeling the probability of an event using the logistic (sigmoid) function. It estimates the likelihood that an observation belongs to a particular class.

It differs from Linear Regression in that linear regression predicts continuous numerical values using a straight-line equation, while logistic regression predicts probabilities that are mapped to class labels. Linear regression can produce values outside the 0‚Äì1 range, whereas logistic regression restricts outputs between 0 and 1. Additionally, logistic regression uses maximum likelihood estimation, while linear regression uses the method of least squares.

**Question 7: Name and briefly describe three common evaluation metrics for regression models.**

**Answer:** --> Three common evaluation metrics for regression models are:

Mean Absolute Error (MAE):
It measures the average absolute difference between the actual and predicted values, providing a clear interpretation of average prediction error.

Mean Squared Error (MSE):
It calculates the average of the squared differences between actual and predicted values, giving more weight to larger errors.

R-squared (R¬≤):
It represents the proportion of variance in the dependent variable explained by the model, indicating how well the model fits the data.

**Question 8: What is the purpose of the R-squared metric in regression analysis?**

**Answer:** --> The purpose of the R-squared (R¬≤) metric in regression analysis is to measure how well the regression model explains the variability of the dependent variable. It indicates the proportion of total variance in the outcome variable that is explained by the independent variable(s) in the model. An R-squared value closer to 1 means the model provides a better fit to the data, while a value closer to 0 indicates poor explanatory power.

**Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.**
**(Include your Python code and output in the code box below.)**

**Answer:** -->

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 6, 8, 10])

# Fit Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

**Question 10: How do you interpret the coefficients in a simple linear regression model?**


**Answer:** --> In a simple linear regression model, the coefficients are interpreted as follows:

Intercept (Œ≤‚ÇÄ):
It represents the expected value of the dependent variable (Y) when the independent variable (X) is zero. It provides a baseline level of Y.

Slope (Œ≤‚ÇÅ):
It indicates the average change in the dependent variable (Y) for a one-unit increase in the independent variable (X). A positive slope means Y increases as X increases, while a negative slope means Y decreases as X increases.

Overall, the coefficients explain the direction, strength, and nature of the relationship between the independent and dependent variables.
