Define Power BI and What are the key components of the Power BI ecosystem? Briefly
explain:
● Power BI Desktop
● Power BI Service
● Power BI Mobile
● Power BI Gateway


**Power BI** is a business analytics tool developed by Microsoft that allows users to connect to different data sources, transform data, and create interactive reports and dashboards for better decision-making.

**Power BI Desktop** is a Windows-based application where users import data, clean and transform it using Power Query, and design interactive reports with charts and visuals.

**Power BI Service** is a cloud-based platform that allows users to publish reports created in Desktop, share them with others, and collaborate in real time through dashboards and workspaces.

**Power BI Mobile** is a mobile application available on smartphones and tablets that enables users to view and interact with dashboards and reports anytime and anywhere.

**Power BI Gateway** is a bridge that connects on-premises data sources (like local servers or databases) to the Power BI Service, allowing secure data refresh and access without moving the data to the cloud.


Question 2 :
Compare the following Power BI visuals:
● Pie Chart vs Donut Chart
● Bar Chart vs Column Chart
When would you prefer one over the other? Give one example for each pair.

Pie Chart vs Donut Chart

A Pie Chart is used to show how different categories contribute to a whole, and it is best when you have a small number of categories with clear differences.

A Donut Chart is similar to a pie chart but has a hole in the center, which makes it better for displaying additional information like total values or labels in the middle.

You should prefer a Pie Chart when you want a simple and clear representation of proportions.

You should prefer a Donut Chart when you want a more modern look or need to display extra information in the center.

Example:
If you are showing market share of 4 companies, a Pie Chart works well for clarity, while a Donut Chart is useful if you also want to show the total market size in the center.


Bar Chart vs Column Chart

A Bar Chart displays data horizontally and is useful when category names are long or when comparing many categories.

A Column Chart displays data vertically and is commonly used to show changes over time or trends.

You should prefer a Bar Chart when you have many categories or long labels because it improves readability.

You should prefer a Column Chart when you want to show time-based data or trends clearly.

Example:
If you are comparing sales across 10 different product names, a Bar Chart is better, while if you are showing monthly sales trends, a Column Chart is more suitable.

Question 3 :
Explain the significance of:
● Star schema vs Snowflake schema
● Primary key vs Foreign key in relationships (Power BI)
Why is cardinality important?

Star Schema vs Snowflake Schema

A Star Schema is a data model where a central fact table is directly connected to multiple dimension tables, making it simple and fast for querying and reporting.

A Snowflake Schema is a more complex model where dimension tables are further normalized into multiple related tables, reducing data redundancy but making queries more complex.

The significance of a Star Schema is that it provides better performance and is easier to understand in Power BI.

The significance of a Snowflake Schema is that it saves storage space and maintains data consistency but may slow down performance.

Primary Key vs Foreign Key (in Power BI)

A Primary Key is a column that uniquely identifies each record in a table, and it does not contain duplicate or null values.

A Foreign Key is a column in one table that links to the primary key in another table, helping to create relationships between tables.

In Power BI, the primary key is usually on the “one” side of a relationship, while the foreign key is on the “many” side.

These keys are important because they allow Power BI to connect tables and enable proper filtering and data modeling.

Why Cardinality is Important

Cardinality defines the type of relationship between two tables, such as one-to-one, one-to-many, or many-to-many.

It is important because it controls how data is filtered and how calculations behave in Power BI.

Correct cardinality ensures accurate results, while incorrect cardinality can lead to wrong aggregations, duplicate values, or misleading reports.

For example, a one-to-many relationship ensures that one customer can have multiple orders, which reflects real-world data correctly.

Question 4 :
Differentiate between:
● Calculated column vs Measure
Also, define Row context and Filter context with simple examples.


Calculated Column vs Measure

A Calculated Column is a column created in a table using a DAX formula, and it is calculated row by row when the data is loaded or refreshed.

A Measure is a calculation that is performed dynamically based on the filters applied in a report, and it is not stored in the table.

A Calculated Column uses row-level data and increases the size of the dataset because values are stored.

A Measure uses aggregated data and does not increase the dataset size because it is calculated only when needed.

You should use a Calculated Column when you need a fixed value for each row, such as creating a “Profit = Sales – Cost” column.

You should use a Measure when you need dynamic calculations, such as “Total Sales = SUM(Sales)” that changes based on filters.

Row Context

Row context means that the calculation is applied to each individual row of a table.

It is commonly used in calculated columns where Power BI evaluates one row at a time.

Example:
If you create a calculated column Profit = Sales – Cost, Power BI calculates profit separately for each row.

Filter Context

Filter context refers to the filters applied to data before performing a calculation, such as slicers, filters, or visuals.

It is mainly used in measures where results change based on selected conditions.

Example:
If you create a measure Total Sales = SUM(Sales) and apply a filter for the year 2024, the measure will only calculate sales for 2024.

Question 5:
What is the difference between a report and a dashboard in Power BI?


A Report in Microsoft Power BI is a detailed, multi-page document that contains multiple visuals such as charts, tables, and graphs, and allows users to explore data with filters and slicers.

A Dashboard in Power BI is a single-page view that displays key metrics and visuals (called tiles) to provide a quick summary of important insights.

A Report is more interactive and is used for deep analysis, while a Dashboard is used for quick monitoring of business performance.

A Report is usually created in Power BI Desktop, whereas a Dashboard is created and shared in the Power BI Service.

Example:
A sales report may contain multiple pages showing region-wise, product-wise, and monthly analysis, while a dashboard will show only key KPIs like total sales, profit, and top-performing region on one screen.
